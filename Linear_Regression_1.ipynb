```python
import numpy as np
import pandas as pd
from numpy import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
cloud= pd.read_excel(r'C:\Users\id\Downloads\cloud gaming dataset.xlsx',usecols=[2,4,5,7,8,6])
```

```python
cloud.head()
```

```python
y = pd.DataFrame(cloud['QoE'])
```

```python
y.head()
```

```python
x=pd.DataFrame()
x['GameType'],x['Framerate'],x['Bitrate'],x['Fluidity'],x['Graphics']=cloud['GameType'],cloud['Framerate'],cloud['Bitrate'],cloud['Fluidity'],cloud['Graphics']
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
pred=regressor.predict(x_test)
```

```python
pred
```

```python
y_pred=pd.DataFrame(pred)
```

```python
y_pred.head()
```

```python
linear_reg=pd.DataFrame()
linear_reg=x_test
```

```python
linear_reg['Actual QoE']= y_test['QoE']
```

```python
linear_reg.reset_index(inplace=True)
```

```python
linear_reg.drop("index", axis=1, inplace=True)
```

```python
linear_reg['Predicte QoE']= y_pred[0]
```

```python
linear_reg.head()
```

```python
linear_reg['Error']= linear_reg['Actual QoE']- linear_reg['Predicte QoE']
```

```python
plt.figure(figsize=(15,8))
plt.plot(linear_reg['Predicte QoE'], 'r')
plt.plot(linear_reg['Actual QoE'])
plt.show()
```
